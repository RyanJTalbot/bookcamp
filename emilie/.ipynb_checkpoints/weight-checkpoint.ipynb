```python
import datetime
import random
import matplotlib.pyplot as plt

```

```python
# make up some data
x = df['date_time']
y = df['Quantity']

# plot
plt.plot(x,y)



# beautify the x-labels
plt.gcf().autofmt_xdate()

# Label plit
plt.xlabel('Date Time') 
plt.ylabel('Weight in Pounds') 


plt.show()
```
