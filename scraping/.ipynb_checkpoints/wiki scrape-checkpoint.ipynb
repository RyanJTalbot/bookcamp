```python
from bs4 import BeautifulSoup
import requests
 
# get URL
page = requests.get("https://en.wikipedia.org/wiki/2006_FIFA_World_Cup_qualification_(UEFA)#Ranking_of_runners-up")
 
# scrape webpage
soup = BeautifulSoup(page.content, 'html.parser')
 
# display scrapped data
print(soup.prettify())

```

```python
# scrape webpage
soup = BeautifulSoup(page.content, 'html.parser')
 
list(soup.children)
 
# find all occurrence of p in HTML
# includes HTML tags
print(soup.find_all('table'))
 
print('\n\n')
 
# return only text
# does not include HTML tags
tables = (soup.find_all('table')[0].get_text())

print(tables.prettify())
```

```python
# create object
object = soup.find(div="mw-parser-output")
 
# find tags
items = object.find_all(class_="wikitable")
result = items[0]
 
# display tags
print(result.prettify())
```
