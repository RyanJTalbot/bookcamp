```python
import requests
url='https://portalseven.com/lottery/powerball_winning_numbers.jsp?fromDate=2003-11-08&toDate=2022-11-06&viewType=3'
response = requests.get(url)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.title)
```

```python
nums = []
blog_titles = soup.findAll('table', attrs={"class":"table table-bordered table-condensed table-striped text-center table-hover"})
for title in blog_titles:
#     nums.append(title.text)
    print(title.text)

# Output:
# Prints all blog tiles on the page
```

```python
nums = []
blog_titles = soup.findAll('tr')
for title in blog_titles:
    nums.append(title.text)
    print(title.text)

# Output:
# Prints all blog tiles on the page
```
