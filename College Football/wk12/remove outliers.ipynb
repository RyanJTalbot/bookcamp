```python
nums = []

for i in times:
    i = str(i)
    nums.append(i[:-2])
```

```python
hours = []
mins = []

for i in nums:
    if(len(i) == 5):
        hours.append(i[0])
        mins.append(i[2:4])        
    else:
        hours.append(i[0:2])
        mins.append(i[3:5])
```
