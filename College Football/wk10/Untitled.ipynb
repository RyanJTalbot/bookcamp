```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('lsu.csv')
```

```python
df
```

```python
res = df['results']
```

```python
res
```

```python
lsu_w = []

for i in res:
    if (i[0] == 'L'):
        lsu_w.append(0)
    else: 
        lsu_w.append(1)
```

```python
lsu_w
```

```python
df1 = df.copy()
```

```python
df1['results'] = lsu_w
```

```python
df1
```

```python
fig, ax = plt.subplots(figsize=(20,5))         
sns.heatmap(df1.corr().loc[['results'],:], linewidth=.9, ax=ax)
```
