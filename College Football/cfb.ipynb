```python
# 0 for HOME Win
# 1 for Home Loss
# 3 for play on neutral site
home_results = []
for index, row in new_df.iterrows():
    if (row['game_location'] == row['Winner']):
        home_results.append(0)
    elif (row['game_location'] == row['Loser']):
        home_results.append(1)
    else:
        home_results.append(row[3])
        
```
