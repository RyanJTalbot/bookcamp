```python
lost = df['Loser']
new_lose = []

for i in lost:
    if(i[2] == ")"):
        new_lose.append(i[4:])
#         if(i[1] == "("):
#             new_winner.append(i[5:])
    elif(i[3] == ")"):
        new_lose.append(i[5:])
    else:
        new_lose.append(i)
    
```

```python
df['losers'] = new_lose
```

```python
location = df['Unnamed: 6']
home = []

for index, row in df.iterrows():
    location = (row['Unnamed: 6'])
    if (location == 'N'):
        home.append('Neutral')
    elif (location == '@'):
        home.append(row['Loser'])
    else:
        home.append(row['Winner'])
```

```python
df['home_team'] = home
```

```python
away = []
for index, row in df.iterrows():
    if (row['home_team'] == row['Winner']):
        away.append(row['Loser'])
    elif (row['home_team'] == row['Loser']):
        away.append(row['Winner'])
    else:
        away.append('neutral')
```

```python
df['away'] = away
```

```python
dates = df1["Date"]
```

```python
month_num = []

for i in dates:
    if(i[1] =='-'):
        if(i[2:5] == 'Sep'):
            month_num.append('9')
        elif(i[2:5] == 'Oct'):
            month_num.append('10')
        elif(i[2:5] == 'Nov'):
            month_num.append('11')
        elif(i[2:5] == 'Aug'):
            month_num.append('8')
    elif(i[2] =='-'):
        if(i[3:6] == 'Aug'):
            month_num.append('8')
        elif(i[3:6] == 'Sep'):
            month_num.append('9')
        elif(i[3:6] == 'Oct'):
            month_num.append('10')
        elif(i[3:6] == 'Nov'):
            month_num.append('11')
```

```python
days = []

for i in dates: 
    if(i[1] == '-'):
        days.append(i[0])
    if(i[2] == "-"):
        days.append(i[0:2])

        
```

```python
years = []

for i in dates: 
    years.append('2022')
```
