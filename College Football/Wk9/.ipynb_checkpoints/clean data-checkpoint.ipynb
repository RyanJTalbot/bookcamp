```python
df = df.drop(columns=['Rk', 'Notes'])
```

```python
df.shape
```

```python
location = df['Unnamed: 4']
```

```python
home = []

for index, row in df.iterrows():
    location = (row['Unnamed: 4'])
    if (location == 'N'):
        home.append(0)
    elif (location == '@'):
        home.append(row['Loser'])
    else:
        home.append(row['Winner'])

```

```python
df['home_team'] = home
```

```python
df
```

```python
away = []
for index, row in df.iterrows():
    if (row['home_team'] == row['Winner']):
        away.append(row['Loser'])
    elif (row['home_team'] == row['Loser']):
        away.append(row['Winner'])
    else:
        away.append(3)
```

```python
df['away'] = away
```

```python
df
```

```python
df.shape
```

```python
homes = df['home_team']
```

```python
df = df.drop(columns=['Unnamed: 4'])
```

```python
df
```

```python
new_df = df.copy()
```

```python
home_results = []
for index, row in new_df.iterrows():
    if (row['home_team'] == row['Winner']):
        home_results.append(0)
    elif (row['home_team'] == row['Loser']):
        home_results.append(1)
    else:
        home_results.append(3)
```

```python
new_df.shape
```

```python
new_df
```

```python
new_df['home_results'] = home_results
```

```python
new_df
```

```python
wins = new_df['Winner']
```

```python
winners = []
for i in wins:
    if (i[2] == ")"):
        winners.append(i[1:2])
        if(i[1] == "("):
            winners.append(i[1:3])
    else:
        winners.append(99)
    
```

```python
len(winners)
```

```python
new_df['win_rank'] = winners
```

```python
lose = new_df['Loser']
```

```python
losers = []
for i in lose:
    if (i[2] == ")"):
        losers.append(i[1:2])
        if(i[1] == "("):
            losers.append(i[1:3])
    else:
        losers.append(99)
```

```python
len(losers)
```

```python
new_df['lose_rank'] = losers
```

```python
new_df
```

```python
new_df = new_df.drop(columns=['Winner', 'Loser', 'home_team', 'away', 'Unnamed: 4'])
```

```python
new_df
```
