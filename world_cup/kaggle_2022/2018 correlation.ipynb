```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.heatmap(df.corr());
```

```python

plt.figure(figsize=(108, 48))
           
# Store heatmap object in a variable to easily access it when you want to include more features (such as title).
# Set the range of values to be displayed on the colormap from -1 to 1, and set the annotation to True to display the correlation values on the heatmap.           
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True)

# Give a title to the heatmap. Pad defines the distance of the title 
#from the top of the heatmap.
        
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':52}, pad=2);
```

```python
plt.figure(figsize=(108, 42))
# define the mask to set the values in the upper triangle to True
mask = np.triu(np.ones_like(df.corr(), dtype=np.bool))

heatmap = sns.heatmap(df.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':26}, pad=8);
```
