```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('wc-7.csv')
```

```python
df
```

```python
counts = df['country']
```

```python
yearz = df['year']
```

```python
data = {'country': counts,
        'years': yearz,}
  
# Create DataFrame
dfs = pd.DataFrame(data)
```

```python
dfs
```

```python
winners = ['france', 'germany', 'spain', 'italy', 'brazil', 
            'argentina', 'england', 'uruguay', ]
```

```python
previous_champs = []

for index, row in dfs.iterrows():
    for j in winners:
        if (j == row['country']):
            previous_champs.append(row['country'])
        
```

```python
previous_champs
```

```python
df['previous_champs'] = previous_champs
```
