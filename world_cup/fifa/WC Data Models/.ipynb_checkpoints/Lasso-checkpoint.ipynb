```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
df = pd.read_csv('wc-wo-outliers.csv')
```

```python
df2 = pd.read_csv('wc-w-outliers.csv')
```

```python
df
```

```python
y2 = df2['results']
X2 = df2.drop(columns=['results'])

X_train2, X_test2, y_train2, y_test2 = train_test_split(
     X2, y2, test_size=0.20, random_state=1)
```

```python
y = df['results']
X = df.drop(columns=['results'])

X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.20, random_state=1)
```

```python
from sklearn import linear_model
```

```python
model_lasso = linear_model.Lasso(alpha=0.01)
model_lasso.fit(X_train, y_train) 
pred_train_lasso= model_lasso.predict(X_train)
```

```python
model_lasso2 = linear_model.Lasso(alpha=0.01)
model_lasso2.fit(X_train2, y_train2) 
pred_train_lasso2= model_lasso2.predict(X_train2)
pred_test_lasso2= model_lasso2.predict(X_test2)
```

```python


print('LASSO REGRESSION')
print('--------------------')
print('RMSE: '+str(np.sqrt(mean_squared_error(y_test2,pred_test_lasso2))))
print('R Squared: '+str(r2_score(y_test2, pred_test_lasso2)))
print('MSE: '+str(mean_squared_error(y_test2, pred_test_lasso2)))
print('MAE: '+str(mean_absolute_error(y_test2, pred_test_lasso2)))
print('MAPE: '+str(mean_absolute_percentage_error(y_test2, pred_test_lasso2)))

print('--------------------')
pred_test_lasso= model_lasso.predict(X_test)

print('RMSE: '+str(np.sqrt(mean_squared_error(y_test,pred_test_lasso)))) 
print('R Squared: '+str(r2_score(y_test, pred_test_lasso)))
print('MSE: '+str(mean_squared_error(y_test, pred_test_lasso)))
print('MAE: '+str(mean_absolute_error(y_test, pred_test_lasso)))
print('MAPE: '+str(mean_absolute_percentage_error(y_test, pred_test_lasso)))
```
