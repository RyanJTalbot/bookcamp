```python
from sklearn.linear_model import LinearRegression
from sklearn.inspection import permutation_importance
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge

```

```python
df.head()
```

```python
y = df['results']
```

```python
X = df.drop(columns=['results'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.20, random_state=42)
```

```python
model = Ridge(alpha=1e-2).fit(X_train, y_train)
model.score(X, y)
```

```python
r = permutation_importance(model, X, y, n_repeats=30, random_state=0)
```

# Permutation feature importance

```python
scoring = ['r2', 'neg_mean_absolute_percentage_error', 'neg_mean_squared_error']
r_multi = permutation_importance(
     model, X, y, n_repeats=30, random_state=0, scoring=scoring)

for metric in r_multi:
     print(f"{metric}") 
     r = r_multi[metric]
     for i in r.importances_mean.argsort()[::-1]:
            if r.importances_mean[i] - 2 * r.importances_std[i] > 0:
                print(f"    {X.columns[i]:<8}"
                      f"{r.importances_mean[i]:.3f}"
                      f" +/- {r.importances_std[i]:.3f}")
                
                
```

- NMAPE = the higher the better
- NMSE
- R2 =  is a statistical measure of how close the data are to the fitted regression line . It is the percentage of variation (from 0 to 1) explained by the relationship between two variables.. metric that is used to evaluate the performance of a regression-based machine learning model.
