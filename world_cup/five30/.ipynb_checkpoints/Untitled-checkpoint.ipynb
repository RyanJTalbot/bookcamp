```python
new_df = df.iloc[16:, :]
```

```python
games = new_df.drop(columns=['date', 'league_id', 'league', 'spi1', 'spi2'])
```

```python
games
```

```python
winner = []
diff = []

for index, row in games.iterrows():
    if (row["proj_score1"] > row['proj_score2']):
        winner.append(row['team1'])
        nums1 = row["proj_score1"] - row['proj_score2']
        diff.append(round(nums1, 2))
    else:
        winner.append(row['team2'])
        nums2 = row["proj_score2"] - row['proj_score1']
        diff.append(round(nums2, 2))
```

```python
winner
```

```python
diff
```

```python
games['winner'] = winner
```

```python
games['margin'] = diff
```
